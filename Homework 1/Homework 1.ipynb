```julia
# 1.4
using Plots             # For plotting
pyplot()

beta=2
A= [0.02 0
    0   0.02]

f(x) = exp(beta*transpose(x)*A*x)
## Plot results
n = 1000
x1 = LinRange(-3,3,n)
x2 = LinRange(-3,3,n)

## Original function values
z  = [f([x1[i], x2[j]]) for j = 1:n, i = 1:n]

## Plot contours of z = f(x) and set values for which contours to plot
p1 = contour(x1,x2,z, 
        legend  = true,
        #clabels = true,
        #clims   = (0,100),
        #levels  = [0.1;0.6;2;collect(5:10:100)],
        size = (500, 500),
        aspect_ratio = :equal,
        xlabel = "x₁",
        ylabel = "x₂",  
        title  = "Contour of f(x)",
)
savefig(p1, "plots/1.4c_contour.png")
p1
```
