```julia
using JuMP
using Ipopt

model = Model(optimizer_with_attributes(Ipopt.Optimizer, "print_level" => 3))

## Variables
@variable(model, x >= 0)
@variable(model, y >= 0)
@variable(model, u[1:4] >= 0)

## Constraints
@constraint(model, 1 - u[1] + u[2] + u[3] + 2u[4] == 0)
@constraint(model, u[1]*(-x - y + 3) == 0)
@constraint(model, u[2]*(-2x + y) == 0)
@constraint(model, u[3]*(2x + y - 12) == 0)
@constraint(model, u[4]*(-3x + 2y + 4) == 0)
@constraint(model,  -x -  y  +  3 <= 0)
@constraint(model, -2x +  y       <= 0)
@constraint(model,  2x +  y  - 12 <= 0)
@constraint(model, -3x + 2y  +  4 <= 0)

## Objective
@objective(model, Min, x - 4y)

## Set initial values
set_start_value(x, 0)
set_start_value(y, 0)
set_start_value(u[1], 0)
set_start_value(u[2], 0)
set_start_value(u[3], 0)
set_start_value(u[4], 0)

## Solve model
optimize!(model)
status = termination_status(model)
println(status)

obj = objective_value(model)
x   = value.(x)
y   = value.(y)
u   = value.(u)

println("\nObjective: $(obj)\n")
println("x = $x")
println("y = $y")
println("u = $u")
```
